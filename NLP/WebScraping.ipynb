```
soup = BeautifulSoup(r.text,"html5lib")
```

```
print(soup.get_text)
```

```
summaries = soup.find_all("a",class_="card__top")
for summary in summaries:
  title = summary.select_one("h2").get_text().strip()
  print(f"*** {title} ***")
  for desc in description:
    details = desc.select_one("p").get_text().strip()
```

```
summaries_1 = soup.find_all("li",class_="catalog-cards__list__item")
summaries_2 = soup.find_all("div",class_="details__layout")
```

```
title_list=[]
for summary in summaries_1:
  title = summary.select_one("h2").get_text().strip()
  title_list.append(title)

details_list=[]
for summary in summaries_2:
  details = summary.select_one("p").get_text().strip()
  details_list.append(details)
```

```
a=dict(zip(title_list,details_list))
```

```
a
```

```
import pandas as pd
df = pd.DataFrame.from_dict(a, orient='index')
```

```
df = df.reset_index()
df.columns = ['title','description']
```
